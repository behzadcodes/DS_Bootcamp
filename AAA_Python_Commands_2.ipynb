# Matplotlib commands

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
# For JUPYTER only - allows you to see plots inside the jupyter notebook envi
```

```python
x = np.linspace(0,5,11)
y = x**2

plt.plot(x,y,'r--')
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.title('First')
plt.subplot(1,2,2)
plt.plot(y,x,'b')
plt.title('Second')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('LARGER PLOT')
axes1.set_xlabel('X Label')
axes1.set_ylabel('Y Label')
axes2.plot(y,x)
axes2.set_title('SMALLER PLOT')
```

```python
# fig,axes = plt.subplots(2,1,figsize=(8,4)) # figsize=(width,height) in inches - can also add dpi=100
fig,axes = plt.subplots(1,2)
axes[0].plot(x,y)
axes[0].set_title('First Plot')
axes[1].plot(y,x)
axes[1].set_title('Second Plot')
plt.tight_layout() # Fixes the isuuse with xtickes and ytickes being jammed into eachother
fig.savefig('Figure1_p.png',dpi=200)
```

```python
fig = plt.figure()
ax=fig.add_axes([0,0,1,1])

ax.plot(x,x**2, color='blue', label='X Squared', linewidth=3, alpha=0.5, marker='o', markersize=10,
       markerfacecolor='r', markeredgewidth=3) # alpha=transparency 
ax.plot(x,x**3, color='#FF8C00', lw=3, ls='--', label='X Cubed') # Color RGB Hex code # lw=linewidth # ls=linestyle
ax.legend(loc='best')
ax.set_xlim([-0.1,4.1])
ax.set_ylim([-5,100])
```

# Seaborn commands

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set(style="darkgrid") # will show grids on the plot, with a transparent background color
# sns.set(style="ticks") # Will only show ticks on the axis, without grids
# other options style=  'white' , 'whitegrid'
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

Distribution Plots

```python
# HUE is a categorical column and splits the plots by different colors based on the specified HUE
sns.distplot(tips['total_bill'],kde=False,bins=30)  # Distribution plot of a uni-variate set of obs - takes one column of a dataframe
# kde=False removes the Kernel Density Estimation curve from the plot
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg') # kind='hex'  # kind='kde'
sns.pairplot(tips, hue='sex',palette='coolwarm') # Shows JOINTPLOT of all combinations of the numerical data in the dataframe
# for combination of a column with itself, it'll plot its histogram
# hue='sex' is for a categorical column # palette='coolwarm' chooses a colormap
sns.rugplot(tips['total_bill']) # Plots a rug plot
sns.kdeplot(tips['total_bill']) # Plots the Kernel Density Estimation (KDE) plot
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg') # kind='hex'  # kind='kde'
```

```python
sns.pairplot(tips, hue='sex',palette='coolwarm') # Shows JOINTPLOT of all combinations of the numerical data in the dataframe
# for combination of a column with itself, it'll plot its histogram
# hue='sex' is for a categorical column # palette='coolwarm' chooses a colormap
```

Categorical Plots

```python
sns.barplot(x='sex',y='total_bill',data=tips, estimator=np.mean) # Plots mean (by default) of a numerical data in different categories
# the estimator by default is mean, but it can be np.std or anything else # X is a categorical column and Y is a numerical column
sns.countplot(x='sex',data=tips) # Plots the count of elements in each category
sns.boxplot(x='day',y='total_bill',data=tips,hue='sex') # X is a categorical column and Y is a numerical column
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True) # Boxplot plus distribution of data points on the sides
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True,hue='sex',split=True) # plots the data distribution as dots
sns.swarmplot(x='day',y='total_bill',data=tips) # Combining stripplot and violinplot

sns.factorplot(x='day',y='total_bill',data=tips,kind='violin') # kind=specify the type of plot 
```

```python
sns.barplot(x='sex',y='total_bill',data=tips, estimator=np.mean) # Plots mean (by default) of a numerical data in different categories
# the estimator by default is mean, but it can be np.std or anything else
plt.title('TITLE')
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='sex') # X is a categorical column and Y is a numerical column
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True) # Boxplot plus distribution of data points on the sides
```

```python
sns.boxplot(x='day',y='total_bill',data=tips, order=sorted(tips['day'].unique()) ) # Or sorted(tips['day'].dropna().unique())
# order gives the order at which the 'day' column will be plotted
```

```python
flights = sns.load_dataset('flights')
fp = flights.pivot_table(index='month',columns='year',values='passengers')
```

```python
sns.heatmap(fp,cmap='magma',linecolor='white',linewidth=1)
```

```python
sns.clustermap(fp,cmap='magma',linecolor='white',linewidth=1)
# clusters so that the months and years that are more similar to each other are shown next to eachother

# add standard_scale=1 to normalize the data between [0,1]
```

```python
iris = sns.load_dataset('iris')
g = sns.PairGrid(iris) # creates an empty pairplot
g.map_diag(sns.distplot) # maps distribution plots on diagonal pairs
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)

# or # g.map(plt.scatter) # to just map scatter plots on all pairs
```

```python
tips = sns.load_dataset('tips')
g = sns.FacetGrid(data=tips,col='time',row='smoker')
g.map(sns.distplot,'total_bill')

# or # g.map(sns.scatterplot,'total_bill','tip')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',markers=['o','v'],# a linear regression model
          scatter_kws={'s':100}) # scatter_kws passes in a set of matplotlib customizations
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex', # you can specify col, row, and hue at the same time
          aspect=0.6,size=8) # aspect is the aspect ration of width and height (width/height)
```
