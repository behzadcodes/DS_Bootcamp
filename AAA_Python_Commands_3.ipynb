```python
data = dict(type = 'choropleth',
            locations = ['AZ','CA','NY'],
            locationmode = 'USA-states',
            colorscale= 'Jet',
            text= ['text1','text2','text3'],
            z=[1.0,2.0,3.0],
            colorbar = {'title':'Colorbar Title'})

layout = dict(geo = {'scope':'usa'})

import plotly.graph_objs as go
choromap = go.Figure(data = [data],layout = layout)

iplot(choromap)
```

```python
df = pd.read_csv('2011_US_AGRI_Exports')
df.head()
```

```python
data2 = dict(type = 'choropleth',
            locations = df['code'],
            locationmode = 'USA-states',
            colorscale= 'Portland',
            text= df['text'],
            z= df['total exports'],
            marker = dict(line = dict(color = 'rgb(255,255,255)',width = 2)),# This marker defines the lines between states
            colorbar = {'title':'Millions USD'})
```

```python
layout2 = dict(title = '2011 US Agriculture Exports by State',
              geo = dict(scope='usa',
                         showlakes = True,
                         lakecolor = 'rgb(85,173,240)') )
```

```python
choromap2 = go.Figure(data = [data2],layout = layout2)
```

```python
# https://plot.ly/python/reference/#choropleth
iplot(choromap2)
```

```python
df3 = pd.read_csv('2014_World_GDP')
df3.head()
```

```python
data3 = dict(type = 'choropleth',
            colorscale= 'Portland',
            locations = df3['CODE'],
            text= df3['COUNTRY'],
            z= df3['GDP (BILLIONS)'],
            colorbar = {'title':'GDP in Billions USD'})

layout3 = dict(
    title = '2014 Global GDP',
    geo = dict(
        showframe = False,
        projection = {'type':'natural earth'}) )

choromap3 = go.Figure(data = [data3],layout = layout3)

iplot(choromap3)
```

```python
df4 = pd.read_csv('2014_World_Power_Consumption')
df4.head()
```

```python
data4 = dict(
        type = 'choropleth',
        colorscale = 'Viridis',
        reversescale = True,
        locations = df4['Country'],
        locationmode = "country names",
        z = df4['Power Consumption KWH'],
        text = df4['Country'],
        colorbar = {'title' : 'Power Consumption KWH'} ) 

layout4 = dict(title = '2014 Power Consumption KWH',
                geo = dict(showframe = False,projection = {'type':'natural earth'}) )

choromap4 = go.Figure(data = [data4],layout = layout4)
iplot(choromap4,validate=False)
```

```python
usdf = pd.read_csv('2012_Election_Data')
usdf.head()
```

```python
data5 = dict(type='choropleth',
            colorscale = 'Viridis',
            reversescale = True,
            locations = usdf['State Abv'],
            z = usdf['Voting-Age Population (VAP)'],
            locationmode = 'USA-states',
            text = usdf['State'],
            marker = dict(line = dict(color = 'rgb(255,255,255)',width = 1)),
            colorbar = {'title':"Voting-Age Population (VAP)"} ) 

layout5 = dict(title = '2012 General Election Voting Data',
              geo = dict(scope='usa',
                         showlakes = True,
                         lakecolor = 'rgb(85,173,240)') )

choromap5 = go.Figure(data = [data5],layout = layout5)
iplot(choromap5,validate=False)
```
