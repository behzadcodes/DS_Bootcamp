```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
# For JUPYTER only - allows you to see plots inside the jupyter notebook environment
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
# FUNCTIONAL method of plotting
plt.plot(x,y,'r--',linewidth=5)
```

```python
plt.plot(x,y,'r--')
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
# Object-Oriented plotting
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8]) # [L,B,W,H]

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('LARGER PLOT')
axes2.plot(y,x)
axes2.set_title('SMALLER PLOT')
```

```python
fig,axes = plt.subplots(1,2) # This is same as mannually creating multiple axes as before

for cc in axes:
    cc.plot(x,y)
```

```python
fig,axes = plt.subplots(1,2)
axes[0].plot(x,y)
axes[0].set_title('First Plot')
axes[1].plot(y,x)
axes[1].set_title('Second Plot')
plt.tight_layout() # Fixes the isuuse with xtickes and ytickes being jammed into eachother
```

```python
fig = plt.figure(figsize=(4,2)) # figsize=(width,height) in inches - can also add dpi=100

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes = plt.subplots(2,1,figsize=(8,4))
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

```python
fig = plt.figure()
ax=fig.add_axes([0,0,1,1])

ax.plot(x,x**2, color='blue', label='X Squared', linewidth=3, alpha=0.5, marker='o', markersize=10,
       markerfacecolor='r', markeredgewidth=3) # alpha=transparency 
ax.plot(x,x**3, color='#FF8C00', lw=3, ls='--', label='X Cubed') # Color RGB Hex code # lw=linewidth # ls=linestyle
ax.legend(loc='best')
ax.set_xlim([-0.1,4.1])
ax.set_ylim([-5,100])
```
